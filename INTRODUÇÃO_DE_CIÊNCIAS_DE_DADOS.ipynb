AULA FUNÇÕES

```python
def somar(a, b):
  resultado = a + b
  return resultado

soma_final = somar(10, 5)
print(f"A soma é: {soma_final}")
```

```python
import datetime
def saudar (nome, hora):
  if hora < 12:
    print(f"Bom dia, {nome}")
  elif hora < 18:
    print (f"Boa tarde, {nome}")
  else:
    print(f"Boa noite, {nome}")

n = input("Qual é o seu nome? ")
h = datetime.datetime.now().hour
saudar(n, h)
```

```python
def converter_dolar(real):
  dolar = real / 5
  return dolar

reais = float(input("Quantidade em Reais: "))
dolares = converter_dolar(reais)
print (f"{reais:.2f} reais convertidos em dólares são: {dolares:.2f} dólares")
```

```python
def filtrar_aprovados(lista_aprovados):
  lista = list()
  for nota in lista:
   if nota >= 7:
    lista_aprovados.append(nota)
  return lista_aprovados

notas = list()
for c in range(3):
 notas.append(float(input(f"{c + 1} nota ")))
print(f"Todas as notas; {notas}")
aprovados = filtrar_aprovados(notas)
print(f"A lista com as notas aprovadas: {aprovados}")

```

```python
def verificar_senha(senha):
  if len(senha) >= 8:
    return True
  else:
    return False

s = input("Crie uma senha: ")
if verificar_senha(s):
  print("Senha válida!")
else:
  print("Senha inválida!")
```
